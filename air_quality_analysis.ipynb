Line Chart for all


```
import pandas as pd
import plotly.graph_objects as go

# Load dataset
file_path = "/content/AT.xlsx"  # Adjust if needed
df = pd.read_excel(file_path)

# 🔹 Convert 'Date' and 'Time' into a single 'DateTime' column
df['DateTime'] = pd.to_datetime(df['Date'].astype(str) + ' ' + df['Time'].astype(str), errors='coerce')

# 🔹 Set 'DateTime' as index
df.set_index('DateTime', inplace=True)

# 🔹 Select pollutants to plot
pollutants = {
    'PM2.5': 60,
    'PM10': 100,
    'NO': 80,
    'NO2': 80,
    'NOx': 80,
    'NH3': 400,
    'SO2': 80,
    'CO': 4,
    'Ozone': 180
}

# Create figure
fig = go.Figure()

# Add pollution lines
for pollutant, limit in pollutants.items():
    if pollutant in df.columns:
        fig.add_trace(go.Scatter(x=df.index, y=df[pollutant], mode='lines', name=pollutant))

# Add threshold lines
for pollutant, limit in pollutants.items():
    fig.add_trace(go.Scatter(x=[df.index.min(), df.index.max()], y=[limit, limit],
                             mode='lines', name=f"{pollutant} Limit", line=dict(dash='dash')))

# Customize layout
fig.update_layout(title="Pollution Levels vs Prescribed Standards",
                  xaxis_title="Date",
                  yaxis_title="Pollution Level",
                  hovermode="x unified")

# Show chart
fig.show()
```

```
# ✅ Install necessary library for interactive plots
!pip install plotly

import pandas as pd
import plotly.graph_objects as go
from google.colab import files

# 🔹 Load dataset (Ensure correct encoding and delimiter)
file_path = "/content/New Filtered Data.csv"
df = pd.read_csv(file_path, sep=",", encoding='utf-8')

# 🔹 Convert 'Date' and 'Time' into a single 'DateTime' column
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')  # Convert Date
df['Time'] = pd.to_datetime(df['Time'], format='%H:%M:%S', errors='coerce').dt.time  # Convert Time
df['DateTime'] = pd.to_datetime(df['Date'].astype(str) + ' ' + df['Time'].astype(str), errors='coerce')

# 🔹 Remove missing values
df = df.dropna(subset=['DateTime'])

# 🔹 Set 'DateTime' as index for time-series plotting
df = df.set_index('DateTime')

# ✅ List of Pollutants and Prescribed Ranges
pollutants = {
    'PM2.5': (0, 60), 'PM10': (0, 100), 'NO': (0, 80), 'NO2': (0, 80), 'NOx': (0, 80),
    'NH3': (0, 400), 'SO2': (0, 80), 'CO': (0, 4), 'Ozone': (0, 180),
    'Benzene': (0, 5), 'Toluene': (0, 5), 'Eth-Benzene': (0, 5), 'MP-Xylene': (0, 5),
    'RH': (None, None), 'WS': (None, None), 'WD': (None, None), 'SR': (None, None),
    'BP': (None, None), 'AT': (None, None)  # 'AT' has no prescribed limit
}

# ✅ Create Interactive Dropdown Chart
fig = go.Figure()

# Add traces for each pollutant
for pollutant, (min_val, max_val) in pollutants.items():
    if pollutant in df.columns:
        fig.add_trace(go.Scatter(x=df.index, y=df[pollutant], mode='lines', name=pollutant, visible=False))

# ✅ Set the first pollutant as visible
fig.data[0].visible = True
first_pollutant = list(pollutants.keys())[0]
min_range, max_range = pollutants[first_pollutant]

# ✅ Create dropdown menu
dropdown_buttons = []
for i, (pollutant, (min_val, max_val)) in enumerate(pollutants.items()):
    visibility = [False] * len(pollutants)
    visibility[i] = True  # Show only the selected pollutant

    # ✅ Add prescribed range shading if applicable
    shapes = []
    if min_val is not None and max_val is not None:
        shapes = [dict(type="rect", xref="paper", yref="y", x0=0, x1=1, y0=min_val, y1=max_val,
                       fillcolor="green", opacity=0.2, layer="below", line_width=0)]

    dropdown_buttons.append(
        dict(label=pollutant,
             method="update",
             args=[{"visible": visibility},
                   {"title": f"{pollutant} Levels Over Time", "shapes": shapes}])
    )

# ✅ Update layout
fig.update_layout(
    updatemenus=[dict(active=0, buttons=dropdown_buttons, direction="down", showactive=True)],
    title=f"{first_pollutant} Levels Over Time",
    xaxis_title="Date & Time",
    yaxis_title="Concentration (µg/m³)",
    template="plotly_dark"
)

# ✅ Show Interactive Chart
fig.show()
```

```
# ✅ Install necessary library for interactive plots
!pip install plotly

import pandas as pd
import plotly.graph_objects as go
from google.colab import files

# 🔹 Load dataset (Ensure correct encoding and delimiter)
file_path = "/content/MergeData.csv"
df = pd.read_csv(file_path, sep=",", encoding='utf-8')

# 🔹 Convert 'Date' and 'Time' into a single 'DateTime' column
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')  # Convert Date
df['Time'] = pd.to_datetime(df['Time'], format='%H:%M:%S', errors='coerce').dt.time  # Convert Time
df['DateTime'] = pd.to_datetime(df['Date'].astype(str) + ' ' + df['Time'].astype(str), errors='coerce')

# 🔹 Remove missing values
df = df.dropna(subset=['DateTime'])

# 🔹 Set 'DateTime' as index for time-series plotting
df = df.set_index('DateTime')

# 🔹 Ensure 'Location' column exists (required for location dropdown)
if 'Location' not in df.columns:
    raise ValueError("The dataset must contain a 'Location' column for location filtering.")

# ✅ List of Pollutants and Prescribed Ranges
pollutants = {
    'PM2.5': (0, 60), 'PM10': (0, 100), 'NO': (0, 80), 'NO2': (0, 80), 'NOx': (0, 80),
    'NH3': (0, 400), 'SO2': (0, 80), 'CO': (0, 4), 'Ozone': (0, 180),
    'Benzene': (0, 5), 'Toluene': (0, 5), 'Eth-Benzene': (0, 5), 'MP-Xylene': (0, 5),
    'RH': (None, None), 'WS': (None, None), 'WD': (None, None), 'SR': (None, None),
    'BP': (None, None), 'AT': (None, None)  # 'AT' has no prescribed limit
}

# ✅ Get unique locations
locations = df['Location'].unique()

# ✅ Create Interactive Dropdown Chart
fig = go.Figure()

# Add traces for each pollutant at each location
for location in locations:
    location_df = df[df['Location'] == location]

    for pollutant, (min_val, max_val) in pollutants.items():
        if pollutant in location_df.columns:
            fig.add_trace(go.Scatter(x=location_df.index, y=location_df[pollutant], mode='lines',
                                     name=f"{pollutant} - {location}", visible=False))

# ✅ Set the first pollutant at the first location as visible
fig.data[0].visible = True
first_location = locations[0]
first_pollutant = list(pollutants.keys())[0]
min_range, max_range = pollutants[first_pollutant]

# ✅ Create dropdown menus for pollutants and locations
pollutant_dropdown = []
location_dropdown = []

for i, (pollutant, (min_val, max_val)) in enumerate(pollutants.items()):
    for j, location in enumerate(locations):
        index = i * len(locations) + j
        visibility = [False] * len(pollutants) * len(locations)
        visibility[index] = True  # Show only the selected pollutant at the selected location

        # ✅ Add prescribed range shading if applicable
        shapes = []
        if min_val is not None and max_val is not None:
            shapes = [dict(type="rect", xref="paper", yref="y", x0=0, x1=1, y0=min_val, y1=max_val,
                           fillcolor="green", opacity=0.2, layer="below", line_width=0)]

        pollutant_dropdown.append(
            dict(label=pollutant,
                 method="update",
                 args=[{"visible": visibility},
                       {"title": f"{pollutant} Levels Over Time ({location})<br>"
                                 f"Prescribed Range: {min_val} - {max_val} µg/m³",
                        "shapes": shapes}])
        )

# ✅ Location dropdown functionality
for j, location in enumerate(locations):
    visibility = [False] * len(pollutants) * len(locations)

    # ✅ Show only pollutants for the selected location
    for i in range(len(pollutants)):
        index = i * len(locations) + j
        visibility[index] = True

    location_dropdown.append(
        dict(label=location,
             method="update",
             args=[{"visible": visibility}])
    )

# ✅ Update layout
fig.update_layout(
    updatemenus=[
        dict(
            buttons=pollutant_dropdown,
            direction="down",
            showactive=True,
            x=0.8,  # Top-right corner
            xanchor="right",
            y=1.15,
            yanchor="top"
        ),
        dict(
            buttons=location_dropdown,
            direction="down",
            showactive=True,
            x=0.95,  # Top-right corner
            xanchor="right",
            y=1.15,
            yanchor="top"
        )
    ],
    title=f"{first_pollutant} Levels Over Time ({first_location})<br>"
          f"Prescribed Range: {min_range} - {max_range} µg/m³",
    xaxis_title="Date & Time",
    yaxis_title="Concentration (µg/m³)",
    template="plotly_dark"
)

# ✅ Show Interactive Chart
fig.show()
```

```
import pandas as pd
import dash
from dash import dcc, html, Input, Output
import plotly.graph_objects as go

# Load the dataset
df = pd.read_csv("/content/MergeData.csv")

# Trim spaces from column names
df.columns = df.columns.str.strip()

# Convert "Date & Time" column to proper datetime format
df["Date & Time"] = pd.to_datetime(df["Date"] + " " + df["Time"], format="%d-%m-%Y %H:%M:%S", errors='coerce')
df.dropna(subset=["Date & Time"], inplace=True)

# Unique pollutant names & locations
pollutants = df.columns[4:]
locations = df["Location"].unique()

# Initialize Dash app
app = dash.Dash(__name__)

# Layout
app.layout = html.Div([
    # Header Row (Left: Pollutant Details, Right: Dropdowns)
    html.Div([
        html.Div([
            html.H3("Pollutant Details", style={'marginBottom': '0px'}),
            html.P("PM2.5 Standard: 60 µg/m³", style={'color': 'red', 'marginTop': '0px'}),
        ], style={'width': '50%', 'display': 'inline-block', 'textAlign': 'left', 'paddingLeft': '20px'}),

        html.Div([
            dcc.Dropdown(
                id='pollutant-dropdown',
                options=[{"label": col, "value": col} for col in pollutants],
                value='PM2.5',
                clearable=False,
                style={'width': '250px', 'display': 'inline-block'}
            ),
            dcc.Dropdown(
                id='location-dropdown',
                options=[{"label": loc, "value": loc} for loc in locations],
                value=locations[0],
                clearable=False,
                style={'width': '250px', 'display': 'inline-block', 'marginLeft': '10px'}
            ),
        ], style={'width': '50%', 'display': 'inline-block', 'textAlign': 'right', 'paddingRight': '20px'})
    ], style={'display': 'flex', 'justifyContent': 'space-between', 'alignItems': 'center', 'padding': '10px'}),

    # Graph
    dcc.Graph(id="line-chart")
], style={'backgroundColor': 'white', 'padding': '20px', 'fontFamily': 'Arial'})

# Callback function
@app.callback(
    Output("line-chart", "figure"),
    [Input("pollutant-dropdown", "value"),
     Input("location-dropdown", "value")]
)
def update_graph(selected_pollutant, selected_location):
    filtered_df = df[df["Location"] == selected_location]

    fig = go.Figure()

    # Line chart
    fig.add_trace(go.Scatter(
        x=filtered_df["Date & Time"], y=filtered_df[selected_pollutant],
        mode='lines+markers', line=dict(color='blue'), name=f"{selected_pollutant} Levels"
    ))

    # Standard lines for common pollutants
    pollutant_standards = {
        "PM2.5": 60,
        "PM10": 100,
        "SO2": 40,
        "NO2": 80,
        "CO": 2  # Assuming 2 mg/m³ (adjust as needed)
    }

    # Add standard line if applicable
    if selected_pollutant in pollutant_standards:
        standard_value = pollutant_standards[selected_pollutant]
        fig.add_hline(y=standard_value, line=dict(color='red', dash='dash'),
                      annotation_text=f"{selected_pollutant} Standard ({standard_value} µg/m³)")

    # Layout adjustments
    fig.update_layout(
        title=f"{selected_pollutant} Levels in {selected_location}",
        xaxis_title="Date & Time",
        yaxis_title=f"{selected_pollutant} Concentration (µg/m³)",
        plot_bgcolor="white",
        paper_bgcolor="white",
        font=dict(color="black"),
        xaxis=dict(showgrid=True, gridcolor='lightgray'),  # Grid for X-axis
        yaxis=dict(showgrid=True, gridcolor='lightgray'),  # Grid for Y-axis
    )

    return fig

# Run the app
if __name__ == '__main__':
    app.run_server(debug=True)
```
